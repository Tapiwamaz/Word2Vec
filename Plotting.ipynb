```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Read the CSV data
df = pd.read_csv('Time vs Samples.csv')

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(df['Negative_Samples'], df['Training_Time_Seconds'], 'bo-', linewidth=2, markersize=8)
plt.xlabel('Number of Negative Samples', fontsize=12)
plt.ylabel('Training Time (Seconds)', fontsize=12)
plt.title('Training Time vs Number of Negative Samples', fontsize=14)
plt.grid(True, alpha=0.3)

# Add value labels on each point
for i, (x, y) in enumerate(zip(df['Negative_Samples'], df['Training_Time_Seconds'])):
    plt.annotate(f'{y:.1f}s', (x, y), textcoords="offset points", xytext=(0,10), ha='center')

plt.tight_layout()
plt.show()

# Print the data for reference
print("Data Summary:")
print(df)
```
